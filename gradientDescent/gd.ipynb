```python
import numpy as np
```

# 生成数据

```python
x = np.arange(100)
y = x * 2 + 0.1 * np.random.rand(100) * x
w = np.ones(1)
b = np.zeros(1)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x, y)
# plt.plot(x, x*w+b)
plt.show()
```

# 定义损失函数

```python
def loss_function(x, y, w, b):
    totalError = 0
    for i in range(0, len(x)):
        loss_i = (y[i] - w * x[i] - b)**2
        totalError += loss_i
    return totalError / len(x)
```

# 定义梯度下降算法

```python
def step_gradient(x, y, w_current, b_current, learning_rate):
    N = len(x)
    gradient_w = 0
    gradient_b = 0
    for i in range(0, len(x)):
        gradient_w += -2 / N * x[i] * (y[i] - w_current * x[i] - b_current)
        gradient_b += -2 / N * 1 * (y[i] - w_current * x[i] - b_current)
    new_w = w_current - learning_rate * gradient_w
    new_b = b_current - learning_rate * gradient_b
    return new_w, new_b
```

```python
num_epochs = 100
for i in range(0, num_epochs):
    w, b = step_gradient(x, y, w, b, 0.00001)
    print(f"第{i+1}个epoch，当前loss为{loss_function(x, y, w, b)[0]}")
```

# 查看结果

```python
plt.scatter(x, y)
plt.plot(x, x*w+b)
plt.show()
```
